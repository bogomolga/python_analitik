```python
from sympy import *
from sympy.plotting import plot
import numpy as np
import matplotlib.pyplot as plt

x = Symbol("x")
f = -12*x**4*sin(cos(x)) - 18*x**3+5*x**2 + 10*x - 30

f
```

```python
#1. Определить корни

solveset(f, x, Reals).evalf(3)

#корни уравнения лежать в диапазоне (−∞,∞)
```

```python
#4. Построить график

plot(f, (x, -50, 50))
```

```python
#5. Вычислить вершины (на отрезке -1.5, 1.5)

plot(f, (x, -1.5, 1.5))
```

```python
#6. Определить промежутки, на которых f > 0

solveset(f>0, x, Reals).evalf(3)

#такие промежутки лежать в диапазоне (−∞,∞)
```

```python
#7. Определить промежутки, на которых f < 0

solveset(f<0, x, Reals).evalf(3)

#такие промежутки лежать в диапазоне (−∞,∞)
```
